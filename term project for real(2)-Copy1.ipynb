```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns
df = pd.read_csv("C:/임승우/wine.data")
df
```

```python
df.columns=['class','alcohol','malic acid','ash', 'alcalinity of ash', 'magnesium','total phenols',
            'flavonoids','nonflavonoid phenols','proanthocyanidins','color intensity','hue',
            'OD280/OD315 of diluted wines','proline']
```

```python
df
```

```python
Total = df['total phenols']
Fla = df['flavonoids']

plt.figure(figsize=(8,5))
plt.boxplot([Total,Fla], labels=['Total phenols','Flavonoids'])

plt.xlabel('Phenols & Flavonoids')
plt.ylabel('Value')
plt.title('Phenols and Flavonoids')
plt.show()
```

```python
Non = df['nonflavonoid phenols']
Fla = df['flavonoids']

plt.figure(figsize=(8,5))
plt.boxplot([Non,Fla], labels=['Non Flavonoids','Flavonoids'])

plt.xlabel('Non Flavonoids phenols & Flavonoids')
plt.ylabel('Value')
plt.title('Non Flavonoids phenols and Flavonoids')
plt.show()
```

```python
Alcohol = df['alcohol']
AOA = df['alcalinity of ash']

plt.figure(figsize=(8,5))
plt.boxplot([Alcohol,AOA], labels=['Alcohol','Alcalinity Of Ash'])

plt.xlabel('Alcohol & Alcalinity of ash')
plt.ylabel('Value')
plt.title('Alcohol and Alcalinity Of Ash')
plt.show()
```

```python
class1_phenols = df.loc[df['class']==1,['class','total phenols']]
class2_phenols = df.loc[df['class']==2,['class','total phenols']]
class3_phenols = df.loc[df['class']==3,['class','total phenols']]
```

```python
plt.figure(figsize=(10,6))
plt.hist(class1_phenols['total phenols'], bins=7, alpha=0.5, label='class1')
plt.hist(class2_phenols['total phenols'], bins=7, alpha=0.5, label='class2')
plt.hist(class3_phenols['total phenols'], bins=7, alpha=0.5, label='class3')

plt.xlabel('total')
plt.ylabel('phenols')
plt.title('Total Phenols & Class')
plt.legend()
plt.grid()
plt.show()
```
