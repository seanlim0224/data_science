```python
import pandas as pd
import matplotlib.pyplot as plt

data=pd.read_excel('C:/data_programming/04_data1.xlsx', index_col='area')
data.head()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

plt.figure(figsize=(20,4))
data2020_good = data['2020_good']
plt.plot(data2020_good, color='b',marker='o')

data2019_good = data['2019_good']
plt.plot(data2019_good, color='r',marker='s')
plt.xlabel('Area')
plt.ylabel('Micrometer')
plt.grid()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

plt.figure(figsize=(20,4))
data2020_verybad = data['2020_verybad']
plt.bar(data2020_verybad.index, data2020_verybad,color='b', alpha=0.5)

data2015_verybad = data['2015_verybad']
plt.bar(data2015_verybad.index, data2015_verybad,color='g', alpha=0.5)

plt.title('compare_2015vs2020')
plt.xlabel('Area')
plt.ylabel('Micrometer')
plt.grid()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

plt.figure(figsize=(20,4))

index =np.arange(2)
bar_width=0.15

data2020_verybad = data['2020_verybad']
plt.bar(index, data2020_verybad,color='b', bar_width)

data2015_verybad = data['2015_verybad']
plt.bar(index+bar_width, data2015_verybad,color='g', bar_width)


plt.title('compare_2015vs2020')
plt.xlabel('Area')
plt.ylabel('Micrometer')
plt.grid()
plt.show()
```

```python
import numpy as np
index = np.arange(7)

plt.figure(figsize=(20,4))

data6 = data.loc['Seoul':'Ulsan', '2015_good':'2020_good']
for year in range(2015,2021):
    year = str(year)+'_good'
    chartdata = data6[year]
    plt.bar(index, chartdata, width=0.15, label=year)
    index = index + 0.15
    
plt.xlabel('Area')
plt.ylabel('Micrometer')
plt.xticks(index-0.53, ['Seoul', 'Busan', 'Daegu', 'Incheon', 'Gwangju', 'Daejeon', 'Ulsan'])
plt.title('2015~2020 Fine Dust(pm10) Good Group Bar Graph')
plt.legend()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
data = pd.read_csv("C:/data_programming/04_data2.csv")
data
```

```python
import pandas as pd
import matplotlib.pyplot as plt

versicolor_SepalWidth = data.loc[data.Species=='versicolor', ['Species', 'SepalWidth']]
setosa_SepalWidth = data.loc[data.Species=='setosa', ['Species', 'SepalWidth']]

plt.figure(figsize=(10,6))
plt.hist(versicolor_SepalWidth['SepalWidth'], bins=8, color='r', alpha=0.5, label='versicolor')
plt.hist(setosa_SepalWidth['SepalWidth'], bins=8, color='b', alpha=0.5, label='setosa')

plt.xlabel('Sepal')
plt.ylabel('Width')
plt.title('Iris Setal Group Histogram')
plt.legend()
plt.grid()
plt.show()
```
