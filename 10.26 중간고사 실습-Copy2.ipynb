```python
data6 = df.loc[:,['passenger_count', 'trip_duration', 'month', 'weekday', 'pickup_hour']]
data6
```

```python
import pandas as pd
import matplotlib.pyplot as plt

January = data6.loc[data6.month==1, ['weekday', 'passenger_count']]
Feburary = data6.loc[data6.month==5, ['weekday', 'passenger_count']]

plt.figure(figsize=(10,6))
plt.hist(January['passenger_count'], bins=8, color='b', alpha=0.5, label='JAN')
plt.hist(Feburary['passenger_count'], bins=8, color='r', alpha=0.5, label='FEB')

plt.xlabel('weekday')
plt.ylabel('Passenger Count')
plt.title('Passenger Count per Weekday')
plt.legend()
plt.grid()
plt.show()
```

```python
data6.loc[data6['month']==1,['month']]=0
data6.loc[data6['month']==5,['month']]=1

G1data=data6.loc[data6.month==0,['month','weekday', 'passenger_count']]
G2data=data6.loc[data6.month==1,['month','weekday', 'passenger_count']]

plt.figure(figsize=(10,6))
sns.histplot(G1data['month'],bins=7,alpha=0.5,label='JAN',color='g',kde=True)
sns.histplot(G2data['month'],bins=7,alpha=0.5,label='FEB',color='r',kde=True)

plt.xlabel('weekday')
plt.ylabel('Passenger Count')
plt.title('Passenger Count per Weekday')
plt.legend()
plt.grid()
plt.show()
```
