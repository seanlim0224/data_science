```python
data.loc[data['Group']==0.0,['Group']]='Year0'
data.loc[data['Group']==1.0,['Group']]='Year1'
data.loc[data['Group']==2.0,['Group']]='Year2'
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

sns.countplot(data=data, x="Group")
plt.xlabel("Group")
plt.ylabel("Count")
plt.title("Label Count")
```

```python
data.loc[data['Group']=='Year0',['Group']]=0
data.loc[data['Group']=='Year1',['Group']]=1
data.loc[data['Group']=='Year2',['Group']]=2
```

```python
G1data=data.loc[data.Group==0,['Diameter','Height','Whole_Weight','Viscra_Weight','Shell_Weight']]
G2data=data.loc[data.Group==1,['Diameter','Height','Whole_Weight','Viscra_Weight','Shell_Weight']]
G3data=data.loc[data.Group==2,['Diameter','Height','Whole_Weight','Viscra_Weight','Shell_Weight']]
```

```python
plt.figure(figsize=(10,6))
sns.histplot(G1data['Diameter'],bins=7,alpha=0.5,label='Year0',kde=True)
sns.histplot(G2data['Diameter'],bins=7,alpha=0.5,label='Year1',color='r',kde=True)
sns.histplot(G3data['Diameter'],bins=7,alpha=0.5,label='Year2',color='g',kde=True)

plt.xlabel('Diameter')
plt.ylabel('count')
plt.title('Abaloneage Diameter - KDE(Kernel Density Estimator)')
plt.legend()
plt.grid()
plt.show()
```

```python
plt.figure(figsize=(10,6))
plt.title('Abaloneage Diameter Seaborn Box Plot Graph')

sns.boxplot(x='Group',y='Diameter',hue='Group',data=data)
plt.show()
```

```python
plt.figure(figsize=(10,6))
plt.title('Abaloneage Height Seaborn Box Plot Graph')

sns.boxplot(x='Group',y='Height',hue='Group',data=data)
plt.show()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_excel('C:/data_programming/05_data1.xlsx')
df
```

```python
data6 = df.loc[:,['gender','age_code','weight', 'height','blood_sugar']]
```

```python
data6.loc[data6['gender']==1,['gender']]='M'
data6.loc[data6['gender']==2,['gender']]='F'
```

```python
data6.loc[df['blood_sugar']>120,['blood_sugar']]='Over'
data6.loc[df['blood_sugar']<120 ,['blood_sugar']]='Under'
```

```python
blood_sugar = data6.groupby(['gender','blood_sugar'])['blood_sugar'].count()
blood_sugar = blood_sugar.to_frame(name='count')
blood_sugar = blood_sugar.reset_index()
```

```python
fig = plt.figure(figsize=(12,12))

area1 = fig.add_subplot(2,1,1)

ax1=sns.barplot(x='gender',y='count',hue='blood_sugar',data=blood_sugar,ax=area1)


fig.suptitle('2020 Health Check Blood Sugar',fontweight='bold')
area1.set_title('Blood Sugar')

plt.show()
```
