```python
plt.figure(figsize=(10,6))
sns.histplot(df['malic acid'],bins=5, alpha=0.5, label='malic acid', kde=True)
sns.histplot(df['ash'],bins=5, alpha=0.5, label='ash', color='r',kde=True)

plt.xlim(1,6)
plt.xlabel('acid&ash')
plt.ylabel('count')
plt.title('KDE')
plt.legend()
plt.grid()
plt.show()
```

```python
plt.figure(figsize=(13,10))
plt.title('14x14Heat Map Graph')

correlation_df=df.corr()
upp_mat = np.triu(correlation_df)

sns.heatmap(correlation_df, annot=True, cmap='RdYlGn', mask=upp_mat)
```
